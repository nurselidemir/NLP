metin temizleme

```python
from nltk.tokenize import TreebankWordTokenizer
from nltk.corpus import stopwords
import string
import re
```

```python
stop_words = set(stopwords.words("english"))
tokenizer = TreebankWordTokenizer()
```

```python
def clean_text(text):
    # Küçük harfe dönüştür
    text = text.lower()
    
    # URL'leri kaldır
    text = re.sub(r"http\S+|www.\S+", "", text)
    
    # Rakamlar ve özel karakterleri temizle (yalnızca harf ve boşluk bırak)
    text = re.sub(r"[^a-z\s]", "", text)
    
    # Tokenleştir (kelimelere ayır)
    words = tokenizer.tokenize(text)
    
    # Stopwords'leri çıkar
    filtered_words = [word for word in words if word not in stop_words]
    
    # Filtrelenmiş kelimeleri birleştir ve döndür
    return " ".join(filtered_words)
```

```python
df2["cleaned_text"] = df2["text"].apply(clean_text)
```

```python
df2
```

vektorizer oluştur

```python
vektorizer = TfidfVectorizer()
X = vektorizer.fit_transform(df2.text)
```

```python
X
```

kelime kümesini incele

```python
feature_names = vektorizer.get_feature_names_out()
```

```python
feature_names
```

```python
tfidf_score = X.mean(axis = 0).A1 
# her kelimenin ortalama tfidf değerleri
```

```python
tfidf_score
```

```python
df_tfidf = pd.DataFrame({"word": feature_names, "tfidf_score": tfidf_score})
# tfidf skorlarını içeren bir df olusturdum.
```
